```python
import matplotlib.pyplot as plt
import numpy as np
import math

list_d_uno_m = [0.0317, 0.0317, 0.0299, 0.029, 0.028, 0.02665, 0.02585, 0.0258, 0.02445, 0.0234, 0.02355, 0.023, 0.0222] # m

list_d_dos = [5.6, 5.48, 5.1, 5.025, 4.79, 4.645, 4.55, 4.2, 4.05, 3.99 , 3.85] #cm
list_d_uno = [3.17,3.12, 2.99, 2.9 , 2.8, 2.665, 2.585, 2.58, 2.435, 2.34, 2.355, 2.3, 2.22] # cm
list_v = [2.6, 2.8, 3.0, 3.2, 3.4, 3.6, 3.8, 4.0, 4.2, 4.4, 4.6, 4.8, 5.0] # kV

error_d = 0.0005
error_y = [error_d] * len(list_d_uno_m)

list_dos_v = [1/math.sqrt(v) for v in list_v]
coef = np.polyfit(list_dos_v, list_d_uno_m, 1) 
m, b = coef

x_fit = np.linspace(min(list_dos_v), max(list_dos_v), 100)
y_fit = m * x_fit + b

#plt.scatter(list_dos_v, list_d_uno_m, label="Datos experimentales", color="blue")

plt.errorbar(list_dos_v, list_d_uno_m, 
             yerr=error_y, fmt='o', 
             color="blue", ecolor="black", capsize=3, 
             label="Datos experimentales")
plt.plot(x_fit, y_fit, 'r--', label=f"Ajuste lineal: y = {m:.4f}x + ({b:.4f})")

plt.xlabel("1/sqrt(V) (V) ")
plt.ylabel("Diámetro 1 (M)")
plt.title("Gráfica diámetro 1 vs 1/sqrt(V)")
plt.legend()
plt.grid(True)
plt.show()
```
